```python
#Extracting trackids and artist ids from each playlist
def getids(playlist_id):
   track_id  = []
   artist_id = []
   album_id = []
   country_list = []
   playlist = sp.playlist(playlist_id)
   for item in playlist['tracks']['items']:
     track = item['track']
     artist = track['artists'][0]['external_urls']['spotify']
     album = track['album']['external_urls']['spotify']
     countries = playlist['name']
     track_id.append(track['id'])
     artist_id.append(artist)
     album_id.append(album)
     country_list.append(countries)
    #  print(country_list)
   return track_id,artist_id,album_id,country_list
```

```python
#function to extract indepth features from the songs
def extractFeatures(id,at,al):
  meta = sp.track(id)
  features = sp.audio_features(id)
  artist_data = sp.artist(at)
  album_data = sp.album_tracks(al)
  # playlist_name = sp.playlist(at)
  

  #meta
  name = meta['name']
  album = meta['album']['name']
  artist = meta['album']['artists'][0]['name']
  if artist_data['genres'] == []:
    genre = 'Missing'
  else:
    genre = artist_data['genres'][0]
  release_date = meta['album']['release_date']
  duration = meta['duration_ms']
  popularity = meta['popularity']
  #country = meta['country']
  # playlists = sp.current_user_playlists()
    

  # song_country = country
  # print(song_country)
  # features
  acousticness = features[0]['acousticness']
  danceability = features[0]['danceability']
  energy = features[0]['energy']
  instrumentalness = features[0]['instrumentalness']
  liveness = features[0]['liveness']
  loudness = features[0]['loudness']
  speechiness = features[0]['speechiness']
  tempo = features[0]['tempo']
  time_signature = features[0]['time_signature']
  valence = features[0]['valence']
  # track_number = album_data[0]['track_number']

  song_features = [name,album,artist,genre,release_date,duration,popularity,acousticness,danceability,
                   energy,instrumentalness,liveness,loudness,speechiness,tempo,time_signature,valence]
  # county = [song_country]
  return song_features
```
